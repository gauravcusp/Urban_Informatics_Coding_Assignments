## Task_1

• Sort data by start_station_id, tripduration

    - Only checking trips with duration <= 3 hours
    
• Only show the top/last 10 records (aka head and tail in SQL)

• List all unique start_station_id values

• Aggregation functions:

    - Count the number of trips (aka wc -l in SQL)
    - Find the average/min/max trip duration

```python
### Sort data by start_station_id, tripduration
### - Only checking trips with duration <= 3 hours
sql_1_1= '''
SELECT * FROM citibike
where tripduration <= 10800 order by start_station_id , tripduration 
'''
queryCartoDB(sql_1_1)
```

```python
### Retrieving the top 10 rows of data and sorting it by start station id and tripduartion. Also data is only for 
### dripduration <= 3 hours
sql_1_2_1='''
SELECT * FROM citibike
where tripduration <= 10800 order by start_station_id asc, tripduration asc
LIMIT 10
'''

queryCartoDB(sql_1_2_1)
```

```python
### Retrieving the last 10 rows of data and sorting descending by start station id and tripduartion. Also data is only for 
### dripduration <= 3 hours
sql_1_2_2='''
SELECT * FROM citibike
where tripduration <= 10800 order by start_station_id desc, tripduration desc
LIMIT 10
'''

queryCartoDB(sql_1_2_2)
```

```python
### selecting all unique start_station_id 
sql_1_3='''
select distinct(start_station_id) as distinct_station_id from citibike
LIMIT 10
'''

queryCartoDB(sql_1_3)
```

```python
### Aggregation functions:
### - Count the number of trips (aka wc -l in SQL)
### - Find the average/min/max trip duration

sql_1_4='''
SELECT distinct start_station_id, 
count(start_station_id) AS tripcount,
avg(tripduration) AS tripavg,
min(tripduration) AS tripmin,
max(tripduration) AS tripmax
FROM citibike
WHERE tripduration <= 10800
GROUP BY start_station_id
ORDER BY start_station_id ASC, tripavg ASC
LIMIT 10

'''

queryCartoDB(sql_1_4)
```

## Task 2

• Selecting trips started on Feb-02-2015 only

• Selecting trips started on the weekends
    - What are average trip duration during weekends?
    
• Can we do the same for weekday?


```python
### Selecting trips started on Feb-02-2015 only

sql_2_1='''
SELECT * FROM citibike
where starttime >= '2015-02-02 00:00'
and starttime < '2015-02-03 00:00'
LIMIT 10
'''
queryCartoDB(sql_2_1)
```

```python
### Selecting trips started on the weekends
sql_2_2_1='''
SELECT * FROM citibike
where extract(DOW FROM starttime) IN (0,6)
LIMIT 10
'''

queryCartoDB(sql_2_2_1)
```

```python
### Selecting trips started on the weekends
### - What are average trip duration during weekends?

sql_2_2_2='''
SELECT start_station_id, starttime,usertype,gender,avg(tripduration) as tripavg FROM citibike
where extract(DOW FROM starttime) IN (0,6) 
group by start_station_id, starttime, usertype, gender
LIMIT 10
'''
queryCartoDB(sql_2_2_2)
```

```python
### Selecting trips started on the weekends
### - What are average trip duration during weekdays?

sql_2_3='''
SELECT start_station_id, starttime,usertype, gender, avg(tripduration) as tripavg FROM citibike
where extract(DOW FROM starttime) NOT IN (0,6) 
group by start_station_id, starttime,usertype, gender
LIMIT 10
'''
queryCartoDB(sql_2_3)
```

## Task 3

• Showing the list of start station locations
    - Using GROUP BY
    
• Showing the number of trips started per station

• … but only for stations within 500m of Time Square!
- The coordinates of Time Square is (40.7577,-73.9857)

```python
### Showing the list of start station locations
### - Using GROUP BY

sql_3_1='''SELECT
 CDB_TransformToWebmercator(CDB_LatLng(
 start_station_latitude,
 start_station_longitude
 )
 ) as the_geom_webmercator,
 MIN(cartodb_id) as cartodb_id
FROM citibike
GROUP BY
 start_station_latitude,
 start_station_longitude
 LIMIT 10
 '''

queryCartoDB(sql_3_1)
```

```python
### Showing the number of trips started per station
sql_3_2='''
SELECT
 start_station_id,
 CDB_TransformToWebmercator(CDB_LatLng(
 start_station_latitude,
 start_station_longitude
 )
 ) as the_geom_webmercator,
 MIN(cartodb_id) as cartodb_id,
 COUNT(tripduration) as numtrips
FROM citibike
GROUP BY
 start_station_id,
 start_station_latitude,
 start_station_longitude
 LIMIT 10
'''

queryCartoDB(sql_3_2)
```

```python
###Showing the number of trips started per station
###• … but only for stations within 500m of Time Square!
###The coordinates of Time Square is (40.7577,-73.9857)

sql_3_3='''
SELECT
 start_station_id,
 CDB_TransformToWebmercator(CDB_LatLng(
 start_station_latitude,
 start_station_longitude
 )
 ) as the_geom_webmercator,
 MIN(cartodb_id) as cartodb_id,
 COUNT(tripduration) as numtrips
FROM citibike
WHERE
ST_DWithin(CDB_LatLng(
 start_station_latitude,
 start_station_longitude
 )::geography,
 CDB_LatLng(40.7577,-73.9857)::geography,
 500)
GROUP BY
 start_station_id,
 start_station_latitude,
 start_station_longitude
 LIMIT 10
'''

queryCartoDB(sql_3_3)
```
